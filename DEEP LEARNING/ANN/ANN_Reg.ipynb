```python
# importing necessary libraries and packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.wrappers.scikit_learn import KerasRegressor
```

```python
# reading data 
data = pd.read_csv('..//input//supercondutor//train.csv')
y = data['critical_temp']
data.drop(['critical_temp'],axis = 1, inplace = True)
```

```python
#divide the data set into categorical & contionus variable
X1 = data.loc[:,data.dtypes==np.float64] #for Continous variables
X2 = data.loc[:,data.dtypes==np.int64]  #for categorical variables
X1.columns
X2['number_of_elements'].unique()
X2['range_Valence'].unique()
X2['range_atomic_radius'].unique()
X21 = pd.get_dummies(X2['number_of_elements'])
X21.head(3)
X22 = pd.get_dummies(X2['range_Valence'])
X22.head(3)
X23 = pd.get_dummies(X2['range_atomic_radius'])
X23.shape
XCAT = pd.concat([X21,X22,X23],axis =1)
```

```python
#pd.DataFrame(XCAT)
mmx = MinMaxScaler()
XCON = pd.DataFrame(mmx.fit_transform(X1),columns=X1.columns)
XCON.head()
```

```python
#combine two individual dataframes
fdata = pd.concat([XCON,XCAT],axis = 1)
```

```python
fdata.shape
fdata.head(5)
```

```python
#Splitingg the dataset'
x_train,x_test,y_train,y_test = train_test_split(fdata,y,test_size = 0.2) 
```

```python
x_train.shape
y_train.shape
```

```python
#creating the neural network
model = Sequential()
model.add(Dense(289, input_dim=289, kernel_initializer='normal', activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
```

```python
history = model.fit(x_train, y_train, epochs=200, batch_size=50,  verbose=1, validation_split=0.2)
```

```python
print(history.history.keys())
# "Loss"
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
