Superconductivity critical temperature prediction 

Import all necessary libraries for model building

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,MinMaxScaler
#from sklearn.scale import MinMaxScaler
```

Read the data set 

```python
data = pd.read_csv("D:\\sai data science\\data science practise\\Data sets\\superconduct\\train.csv")
```

```python
data.head(3)
```

```python
data.info()
```

One thing is clear that there is no null values in the dataset as all columns are having all 21263 notnull values

```python
y = data['critical_temp']
data.drop(['critical_temp'],axis = 1,inplace=True)
```

```python
data.columns
```

```python
data.corr()
```

Just directly built the model and will compare the results of preprocessed data and plain data 

Divide the data set into train and test 

```python
x_train,x_test,y_train,y_test = train_test_split(data,y,test_size = 0.2)
x_train.columns
y_train
```

# 1.Trying with RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from math import sqrt
rfr = RandomForestRegressor()
rfrmodel1 = rfr.fit(x_train,y_train)
rfrmodel1
#make predictions by this model
y_pred = rfrmodel1.predict(x_test)
#checking the model performance by MAE,RMSE,MSE
print("MAE : ", metrics.mean_absolute_error(y_test,y_pred))
print("RMSE : ", sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("MSE : ", metrics.mean_squared_error(y_test,y_pred))
```

# 2.Linear regression model 

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lrmodel1 = lr.fit(x_train,y_train)
#making predictions by lrmodel1
y_predlr1 = lrmodel1.predict(x_test) 
#checking the model performance by MAE,RMSE,MSE
print("MAE : ", metrics.mean_absolute_error(y_test,y_predlr1))
print("RMSE : ", sqrt(metrics.mean_squared_error(y_test,y_predlr1)))
print("MSE : ", metrics.mean_squared_error(y_test,y_predlr1))
```

preprocessing the data with minmaxscaler,get dummies for categorical variables and we will see if there is any improvement in the model performance 

```python
# as we know from data.info()  that there are three categorical variables,we need to convert them into binary variables
noe = pd.get_dummies(data['number_of_elements'])
rar = pd.get_dummies(data['range_atomic_radius'])
rv = pd.get_dummies(data['range_Valence'])
data2 = data
# wealready created dummy variables for those three variables,so we can drop the original columns from the dataframe
data2.drop(['number_of_elements','range_atomic_radius','range_Valence'],axis =1,inplace= True)
#data2 = pd.DataFrame(data2)
```

```python
#add those three dummies variables to our data set
data3 = pd.concat([data2,noe,rv,rar],axis=1)
type(data2)
type(noe)
type(rar)
type(rv)
```

```python
total_rows = len(data3.axes[0])
total_cols = len(data3.axes[1])
total_rows
total_cols
```

```python
#appluy minmax scaling 
cols = data3.columns
mxm = MinMaxScaler()
mxm_data = pd.DataFrame(mxm.fit_transform(data3),columns = cols)
```

```python
#make predictions using new data which is preprocessed and will compare the results
x_tr,x_te,y_tr,y_te = train_test_split(mxm_data,y,test_size = 0.2)
```

# Random forest regressor model 2

```python
rfr2 = RandomForestRegressor()
rfrmodel2 = rfr2.fit(x_tr,y_tr)
y_pred2 = rfrmodel2.predict(x_te)
#checking the model performance by MAE,RMSE,MSE
print("MAE : ", metrics.mean_absolute_error(y_te,y_pred2))
print("RMSE : ", sqrt(metrics.mean_squared_error(y_te,y_pred2)))
print("MSE : ", metrics.mean_squared_error(y_te,y_pred2))
```

```python
# we saw that there is little bit improvement in the model performance , now try to tune the hyper parameters of random forest regressor 
```

# Linear regression model2

```python
lr2 = LinearRegression()
lrmodel2 = lr2.fit(x_tr,y_tr)
y_predlr2 = lrmodel2.predict(x_te)
#checking the model performance by MAE,RMSE,MSE
print("MAE : ", metrics.mean_absolute_error(y_te,y_predlr2))
print("RMSE : ", sqrt(metrics.mean_squared_error(y_te,y_predlr2)))
print("MSE : ", metrics.mean_squared_error(y_te,y_predlr2))
```

```python
# hyper parameter tuning for Random forest model2
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
rf2 = RandomForestRegressor()
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

rf_random = RandomizedSearchCV(estimator = rf2, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
rf_random.fit(x_te, y_te)
# to find the best params
```

```python
rf_random.best_params_
```
